```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Model comparison data focusing on accuracy
model_comparison = {
    "Model": ["Logistic Regression", "Naive Bayes", "SVM", "Random Forest", "XGBoost"],
    "Accuracy": [0.78, 0.765, 0.76, 0.745, 0.715]
}

# Create a DataFrame
df_comparison = pd.DataFrame(model_comparison)

# Identify the model with the best accuracy
best_model = df_comparison.loc[df_comparison['Accuracy'].idxmax()]

# Print the best model
print(f"The model with the best accuracy is: {best_model['Model']} with an accuracy of {best_model['Accuracy']}")

# Set plot size
plt.figure(figsize=(8, 5))

# Create a seaborn barplot for accuracy
sns.barplot(x="Model", y="Accuracy", data=df_comparison)

# Add labels and title
plt.title('Model Accuracy Comparison')
plt.xlabel('Model')
plt.ylabel('Accuracy')

# Highlight the best model in the plot
plt.axhline(y=best_model['Accuracy'], color='r', linestyle='--', label='Best Model: ' + best_model['Model'])

# Rotate x labels for better readability
plt.xticks(rotation=45)

# Add legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
